## Exploring the data
- Gathering a list of all kickers, their names and ID's will be important for connecting them to plays tracking/plays data.
    - There are a total of 47 kickers. I have their names but most importantly their ID's

- I want to get their tracking information for specific game(s). Interesting that if the gameId's match then I know they played against each other or were on same team.

## Getting a list of all kickers

```python
all_k = tracking[tracking['position']=='K']
k_data = all_k[['nflId', 'gameId', 'displayName', 'position', 'team']].copy()
k_data = k_data.drop_duplicates(keep='first')

# Show kicker data
k_data
```

## Understanding how many kickers there are, the value in the tuples are how many games they have played.

```python
temp = k_data.groupby(['displayName', 'nflId']).count()
print("There are a total of:", len(temp), "kickers.")
temp.head(15)
```

## Getting the plays where a field goal attempt was made for kickers in a list

```python
df_play_fga = plays.loc[(plays['kickerId'].isin([46298.0, 52596.0, 46236.0, 35102.0, 40125.0, 47423.0, 38701.0, 45046.0, 52656.0])) & (plays['specialTeamsPlayType']=='Field Goal')]
df_play_fga = df_play_fga.rename(columns={'kickerId': 'nflId'})
df_play_fga
```

## Connecting the kickers we are interested in back to the tracking2020 data to get their x,y positions of field goal attempt.

```python
tracking_fga = tracking[tracking['event']=='field_goal_attempt']
tracking_fga
```

```python
end_fga = tracking[tracking['event']=='field_goal']
end_fga = end_fga[end_fga['displayName']=='football']
end_fga
```

```python
merged_k = df_play_fga.merge(tracking_fga, how='left', on=['nflId', 'gameId', 'playId'])
```

```python
output_k = merged_k.drop(columns=['quarter', 'down', 'yardsToGo', 'possessionTeam', 
                                  'playDescription', 'frameId', 'returnerId', 'kickBlockerId',
                                  'a', 'o', 'dis', 'dir', 'penaltyCodes', 'gameClock', 'penaltyJerseyNumbers',
                                  'time', 'jerseyNumber', 'preSnapHomeScore', 'preSnapVisitorScore', 'playResult', 
                                  'passResult', 'penaltyYards', 'kickReturnYardage'])



# Since the plays data has data from 2018 -> 2022 I an not interested in any data where the merge could not be done.
# This is why I chose a left join merge to maintain data from kickers and only join where the tracking information 
# for a kick is done.
# I then drop all instances where the displayname = Nan becuase this means there was no match for the playid gameid 
# and nflid to the tracking information

output_k = output_k[merged_k['displayName'].notna()]
output_k
```

## Connecting the position of the football when the event says field goal to the data so i have the end position of the football.

```python
output_football_too = output_k.merge(end_fga, how='left', on=['playId'])
output_football_too = output_football_too[output_football_too['time'].notna()]
output_football_too
```

```python
# output_k = output_k.groupby(['displayName', 'nflId', 'gameId'])
# output_k.first()
```

```python
output_football_too = output_football_too.drop(columns=['o', 'dir', 'nflId_y', 'jerseyNumber', 
                                  'position_y'])
```
